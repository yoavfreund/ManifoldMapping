### 2D

```python
diag={}
```

```python
k=20
iter=100000
A=zeros([k+2,k+2])
A[0:2,0:2]=1

for r in range(iter):
    B=zeros([k,k])
    for i in range(-1,2):
        for j in range(-1,2):
            B=B+A[1+i:k+1+i,1+j:k+1+j]
    B=B/9.

    A[1:-1,1:-1]=B
    A[0:2,0:2]=1
    A[-2:,-2:]=0
    A[0,:]=A[1,:]
    A[:,0]=A[:,1]
    A[-1,:]=A[-2,:]
    A[:,-1]=A[:,-2]
    
diag[iter] = [A[i,i] for i in range(1,k+1)]
```

```python
diag.keys()
```

```python
for key in diag.keys():
    plot(diag[key],label=str(key))
title('diagonal for %dX%d matrix'%(k,k))
legend()
grid()
```
