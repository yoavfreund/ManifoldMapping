```python

class CoverTree(object):
    def __init__(self,center,radius,level):
        self.center=center
        self.radius=radius
        self.counter=0  # number of points inserted into this tree
        self.too_far=0  # count how many points were too far (should only be non-zero at root
        self.level=level
        self.children = []
    def insert(self,x):
        d= _dist(x,self.center)
        if d>self.radius:
            self.too_far+=1
            return
        self.counter+=1
        if len(self.children)==0:
            self.children = [CoverTree(x,self.radius/2,self.level+1)]
        else:
            inserted=False
            for child in self.children:
                if _dist(x,child.center)<self.radius/2:
                    child.insert(point)
                    inserted=True
            if not inserted:
                self.children.append(CoverTree(x,self.radius/2,self.level+1))
    def _str_to_level(self,max_level):
        s='  '*self.level+'%3d: r=%4.2f, no_child=%d, count=%d'%(self.level,self.radius,len(self.children),self.counter)+'\n'
        if self.level < max_level and len(self.children)>0:
            for i in range(len(self.children)):
                s+=self.children[i]._str_to_level(max_level)
        return s    
```
